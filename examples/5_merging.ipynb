In this notebook, we show how to merge the ouput `.json` files that are written at inference time. The output is one `.json` file with all the annotations.

```python
from src.merging import CellMerger, GSONTile
from pathlib import Path

# Merge cell annotations
in_dir = "/path/to/the/geojson/files"
result_dir= Path("/path/to/the/resultdir")
fname_cells = "wsi_cells.json"
fn = result_dir / fname_cells

c = CellMerger(in_dir=in_dir)
c.merge(fname=fn)
```

Let's visualize what does a WSI-size nuclei segmentation masks look like

```python
import matplotlib.pyplot as plt

cells = GSONTile(fn, xmin=0, ymin=0)

cells.gdf.plot(
    "class_name",
    figsize=(15, 15),
    categorical=True,
    legend=True,
    legend_kwds={"fontsize":30, "loc": "upper left"}
)
plt.axis('off')
```
